```python
import numpy as np
import pandas as pd
from catboost import CatBoostClassifier, Pool
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.model_selection import GridSearchCV
import pickle
```

```python
data = pd.read_csv('data/features_for4.csv')
```

```python
# Вектор значений целевой переменной
y = data['isFraud']
# Матрица объект-признак
# Удалим колонку TransactionID, т.к. корреляция для TransactionID и TransactionDT близка к 1
X = data.drop(columns=['isFraud', 'TransactionID'])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=123)
```

```python
cat_feat = list(data.describe(include='object').columns)
```

```python
%%time
model = CatBoostClassifier(n_estimators=5000,
                           depth=9,
                            task_type="GPU",
                           learning_rate=0.1,
                           loss_function='Logloss',
                           cat_features=cat_feat,
                           verbose=True)
# train the model
model.fit(X_train, y_train)
# make the prediction using the resulting model
preds_class = model.predict(X_test)
preds_proba = model.predict_proba(X_test)
print("class = ", preds_class)
print("proba = ", preds_proba)
```

### Метрики качества на тестовой выборке

```python
# метрики качества
def get_scores(model_name, y_test, pred, pred_proba):
    scores = pd.DataFrame({'score_name':['accuracy', 'precision', 'recall', 'f1', 'roc-auc'],
                          model_name:[accuracy_score(y_test, pred), precision_score(y_test, pred), 
                                         recall_score(y_test, pred), f1_score(y_test, pred),
                                        roc_auc_score(y_test, pred_proba)]})
    scores.set_index('score_name', inplace=True)
    return scores
```

```python
baseline_pred = np.zeros(y_test.shape[0])
baseline_scores = get_scores('baseline', y_test, baseline_pred, baseline_pred)

full_data_scores = get_scores('full_data', y_test, model.predict(X_test),
                              model.predict_proba(X_test)[:, 1])
```

```python
scores = pd.concat([baseline_scores, full_data_scores], axis=1)
```
