```python
import pandas as pd
```

```python
X_raw = pd.read_csv('clear_and_nan.csv')
```

```python
pd.set_option('display.max_columns', None)
X_raw.describe(include='object')
```

```python
def get_mean_encoding(df, columns, target):
    df_copy = df.copy()
    for col in columns:
        mean_encoded = df.groupby(col)[target].mean()
        df_copy[col] = df_copy[col].map(mean_encoded)
    return df_copy
```

```python
cat_mean_enc_cols = ['card1', 'card2', 'card3', 'card5', 'addr1', 'addr2',
                     'id_13', 'id_17', 'id_19', 'id_20', 'id_31', 'DeviceInfo']
X_mean_encoded = get_mean_encoding(X_raw, cat_mean_enc_cols, 'isFraud')
```

```python
y = X_mean_encoded['isFraud']
X = X_mean_encoded.drop(columns=['isFraud'])

# map binary categorical columns values to 0 or 1
d_for_map = {'F': 0, 'T': 1, 'NotFound': 0, 'Found': 1, 'New': 0, 'desktop': 0, 'mobile': 1}
cat_map_cols = ['M1', 'M2', 'M3', 'M5', 'M6', 'M7', 'M8', 'M9',
           'id_12', 'id_16', 'id_28', 'id_29', 'id_35', 'id_36', 'id_37', 'id_38', 'DeviceType']

for col in cat_map_cols:
    X[col] = X[col].map(d_for_map)

# code One Hot Encoding only cat_dum_cols
cat_dum_cols = ['card4', 'card6','ProductCD', 'P_emaildomain', 'R_emaildomain', 'M4', 'id_15']
X = pd.get_dummies(X, drop_first=True, columns=cat_dum_cols)
```

```python
X.describe()
```

```python
X = X.drop(columns=['TransactionID'])
```

Нормализация

```python
from sklearn import preprocessing

normalizer = preprocessing.MinMaxScaler()
X_real_norm_np = normalizer.fit_transform(X)
X_norm = pd.DataFrame(data=X_real_norm_np)
```

Обученик модели

```python
from sklearn.model_selection import train_test_split

Xtrain, Xtest, ytrain, ytest = train_test_split(X_norm, y, test_size=0.25, random_state=123)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(Xtrain, ytrain)
```

```python
model.coef_
```

```python
pred_mse = model.predict(Xtest)
```

```python
from sklearn.metrics import r2_score

r2_score(ytest, pred_mse)
```

```python
r2_score(ytrain, model.predict(Xtrain))
```

```python
from sklearn.linear_model import Ridge

model_l2 = Ridge(alpha=1)

model_l2.fit(Xtrain, ytrain)

pred_l2 = model_l2.predict(Xtest)

r2_score(ytest, pred_l2)
```

```python
model_l2.coef_
```

```python
r2_score(ytrain, model_l2.predict(Xtrain))
```

```python
from sklearn.linear_model import SGDRegressor

lr_mse = SGDRegressor(loss='squared_loss')

lr_mse.fit(Xtrain, ytrain)

pred_mse = lr_mse.predict(Xtest)

r2_score(ytest, pred_mse)
```
